Data Visualization in Python.

```python
import pandas as pd 
import matplotlib.pyplot as plt

data = {'name':['Arya', 'Bacteria', 'Cells', 'Zid', 'Cell', 'Lilla'],
        'Moblies_sales':[2232, 3223, 2122, 6776, 8989, 2456],
        'TV_sales':[3323, 824, 1213, 2677, 3576, 1456]}
ds = pd.DataFrame(data)
ds
# print('*************')
# print(ds)
```

create a bar plot of the sales volume .

```python
ds.plot(kind = "bar", x = "name", y = "Moblies_sales", color = "blue")
ds.plot(kind = "bar", x = "name", y = "TV_sales", color = "green")
ds.plot.bar( x = "name", y = "TV_sales", color = "green")
# plt.xlabel("Name")
# plt.ylabel("TV_sales & Moblies_sales")
# plt.show()
```

pie chart.

```python
ds.plot.pie(y = "Moblies_sales")
ds.plot.pie(y = "TV_sales")
ds.plot(kind = "pie", y = "TV_sales")
```

```python
ds.plot(kind = "box", x = "name", y = "Moblies_sales")
ds.plot(kind = "box", y = "TV_sales")
```

Create the Area of the items sales.

```python
ds.plot.area()
```

Create the stacked bar plot of item sales.

```python
ds.plot.bar(stacked = True)
```
