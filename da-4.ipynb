### matplotlib.pyplot
matplotlib.pyplot is a state-based interface to matplotlib. It provides an implicit, MATLAB-like, way of plotting. It also opens figures on your screen, and acts as the figure GUI manager.

pyplot is mainly intended for interactive plots and simple cases of programmatic plot generation:

```python
x = np.arange(0, 5, 0.1)
y = np.sin(x)
plt.plot(x, y)
```


The explicit (object-oriented) API is recommended for complex plots, though pyplot is still usually used to create the figure and often the axes in the figure. See .pyplot.figure, .pyplot.subplots, and .pyplot.subplot_mosaic to create figures, and Axes API <../axes_api> for the plotting methods on an axes:



```python
x = np.arange(0, 5, 0.1)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y)
```
